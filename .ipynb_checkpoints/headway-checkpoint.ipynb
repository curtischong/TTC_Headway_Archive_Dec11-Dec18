```python
def graphLoc(name):
    arr = []
    text_file = open("Dec11-Dec18/"+name, "r")
    lines = text_file.read()[2:].split('\n')
    lines = lines[2:]
    for x in lines:
        thearr = x.split("~")
        time = thearr[0]
        thearr = thearr[1:]
        while len(thearr) > 0:
            arr.append([time] + thearr[:5])
            thearr = thearr[5:]

    vehDf = pd.DataFrame(arr, columns=['timestamp','veh_id','dir_id','heading','lat','lon'])
    
    text_file = open("Dec11-Dec18/"+name, "r")
    lines = text_file.read().split("\n")
    lines = lines[1].split("~")

    arr5 = np.array(lines)
    arr5 = np.delete(arr5, np.arange(0, arr5.size, 5))
    arr5 = np.delete(arr5, np.arange(0, arr5.size, 4))
    arr5 = np.delete(arr5, np.arange(0, arr5.size, 3))
    
    arr1 = arr5[0::2]    
    arr2 = arr5[1::2]
    lon = arr2.astype(np.float)
    lat = arr1.astype(np.float)

    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.scatter(vehDf["lat"].astype(np.float), vehDf["lon"].astype(np.float))
    ax.scatter(lat, lon,  c="red")
    plt.title("Route "+name[:-4])
    plt.ylabel('Longitude')
    plt.xlabel('Latitude')
    plt.savefig("locations_on_map/"+name[:-4]+'.png')
```

```python
text_file = open("Dec11-Dec18/5.txt", "r")
lines = text_file.read().split("\n")
lines = lines[1].split("~")
lines[:25]
```

```python
stopId = np.array(lines[0::5])
stopId = stopId[:-1]#get rid of the extra value at the end caused by an extra ~
stopName = np.array(lines[2::5])
stopLat = np.array(lines[3::5])
stopLon = np.array(lines[4::5])
stopsArr = np.column_stack((stopId,stopName,stopLat,stopLon))
stopsDf = pd.DataFrame(stopsArr)
stopsDf.head()
```

```python
headwayArr = []
for stop in stopsDf.iterrows():
    busAtThisTime = []
    lastTime = 0
    lastBus = ""
    for bus in vehDf.iterrows():
        
        if ((math.sqrt((float(bus[1][4]) - float(stop[1][2]))**2 + (float(bus[1][5]) - float(stop[1][3]))**2) < 0.00025) & ((lastBus != str(bus[1][1])) | ((int(bus[1][0])) - lastTime > 60000))):
            busAtThisTime.append(bus)
            lastBus = str(bus[1][1])
            lastTime = int(bus[1][0])
    headwayArr.append([stop[1][0],busAtThisTime])
#headwayArr
```
